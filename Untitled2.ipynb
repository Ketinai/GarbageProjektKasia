```python
import os, shutil, random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

```python
#List of categories

IMAGE_WIDTH = 224    
IMAGE_HEIGHT = 224
IMAGE_SIZE=(IMAGE_WIDTH, IMAGE_HEIGHT)
IMAGE_CHANNELS = 3

data = "/Users/katarzynaiwaszkiewicz/Desktop/GarbageProjektKasia/garbage_classification"
data_classes = os.listdir(data)
data_classes
```

```python
class_counts = {}

for cls in os.listdir(data):
    class_path = os.path.join(data, cls)
    class_counts[cls] = len(os.listdir(class_path))
    
class_counts = dict(sorted(class_counts.items(), key=lambda x:x[1], reverse = True))
class_counts
```

```python
g = sns.barplot(x = list(class_counts.values()), y = list(class_counts.keys()))
```

```python
#List of filenames

file_names = os.listdir(class_path)
random.shuffle(file_names)
length = len(file_names)
train_file_names = file_names[:int(length*0.8)]
test_file_names = file_names[int(length*0.8):]

file_names
```

```python
for classification in os.listdir(data):
    count_class = 0
    for photo in os.listdir(data + '/' + classification):
        count_class += 1
    print(str(classification) + " has " + str(count_class) + " images.")
```

```python
data_frame = pd.DataFrame()
for classification in os.listdir(data):
    temp = pd.DataFrame()
    temp['garbage type'] = classification
    i = 0
    for photo in os.listdir(data + '/' + classification):
        temp.loc[i, 'image path'] = data + classification + "/" + photo
        temp.loc[i, 'garbage type'] = classification
        i += 1
    data_frame = pd.concat([data_frame, temp], ignore_index=True)
    del temp

data_frame.head(10)
```

```python
pd.unique(data_frame['garbage type'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

train_df, validate_df = train_test_split(data_frame, test_size=0.2, train_size=0.8, random_state=42)

# We first split the data into two sets and then split the validate_df to two sets
validate_df, test_df = train_test_split(validate_df, test_size=0.5,  random_state=42)

train_df = train_df.reset_index(drop=True)
validate_df = validate_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)

total_train = train_df.shape[0]
total_validate = validate_df.shape[0]

print('train size = ', total_train , 'validate size = ', total_validate, 'test size = ', test_df.shape[0])
```

```python
train_data_dir =data
batch_size = 128                           
target_size = (224,224)
validation_split = 0.2
```

```python
train= tf.keras.preprocessing.image_dataset_from_directory(
    train_data_dir,
    validation_split=validation_split,
    subset="training",
    seed=50,                                                         
    image_size=target_size,
    batch_size=batch_size,
)
validation= tf.keras.preprocessing.image_dataset_from_directory(
    train_data_dir,
    validation_split=validation_split,
    subset="validation",
    seed=100,                                                         
    image_size=target_size,
    batch_size=batch_size,
)
```

```python
class_names = train.class_names
class_names
```

```python
plt.figure(figsize=(10, 10))
for images, labels in train.take(1):
    for i in range(8):
        ax = plt.subplot(8, 4, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
base_model = tf.keras.applications.EfficientNetV2B1(input_shape=(224,224,3),include_top=False,weights='imagenet')
base_model.trainable = False
keras_model=keras.models.Sequential()
keras_model.add(base_model)
keras_model.add(keras.layers.Flatten()) 
keras_model.add(keras.layers.Dropout(0.5))
keras_model.add(keras.layers.Dense(12,activation=tf.nn.softmax))     #12 classes
keras_model.summary()
```

```python
import keras 
from tqdm import tqdm
from keras.callbacks import EarlyStopping,ModelCheckpoint
from sklearn.metrics import confusion_matrix , accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import glob 
import pandas as pan
import matplotlib.pyplot as plotter
import warnings
warnings.filterwarnings('ignore')
```

```python
checkpoint =ModelCheckpoint("my_cv_model.h5", save_best_only=True)
```

```python
early_stopping =EarlyStopping(patience=10, restore_best_weights=True)
```

```python
keras_model.compile(optimizer ='Adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
hist=keras_model.fit_generator(train,epochs=10,validation_data=validation,callbacks=[checkpoint,early_stopping])
```

```python
score, acc = keras_model.evaluate(validation)
print('Test Loss =', score)
print('Test Accuracy =', acc)
```

```python
hist_=pd.DataFrame(hist.history)
hist_
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(hist_['loss'],label='Train_Loss')
plt.plot(hist_['val_loss'],label='Validation_Loss')
plt.title('Train Loss & Validation Loss',fontsize=20)
plt.legend()
plt.subplot(1,2,2)
plt.plot(hist_['accuracy'],label='Train Accuracy')
plt.plot(hist_['val_accuracy'],label='Validation Accuracy')
plt.title('Train Accuracy & Validation Accuracy',fontsize=20)
plt.legend()
```

```python
X_val,y_val,y_pred=[],[],[]
for images, labels in validation:
    y_val.extend(labels.numpy())
    X_val.extend(images.numpy())
predictions=keras_model.predict(np.array(X_val))
for i in predictions:
    y_pred.append(np.argmax(i))
df=pd.DataFrame()
df['Actual'],df['Prediction']=y_val,y_pred
df
```

```python
plt.figure(figsize=(10,10))
for i in range(32):
    ax = plt.subplot(8, 4, i + 1)
    plt.imshow(X_val[i].astype("uint8"))
    plt.title('Actual: {}      Pred: {}'.format(class_names[y_val[i]],class_names[y_pred[i]]),  color='green', fontsize=7)
    plt.axis("off")
```

```python
ax= plt.subplot()
CM = confusion_matrix(y_val,y_pred)
sns.heatmap(CM, annot=True, fmt='g', ax=ax,cbar=False,cmap='RdBu')
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels') 
ax.set_title('Confusion Matrix')
plt.show()
CM
```

```python
Acc = accuracy_score(y_val,y_pred)
print("accuracy is: {0:.4f}%".format(Acc * 100))
```
